```python
from collatex import *

collation = Collation()

temoin_MS4A3a = open( "textes/missy/MS4A3a.txt", encoding='utf-8' ).read()
temoin_MS4A1a = open( "textes/missy/MS4A1a.txt", encoding='utf-8' ).read()
temoin_cendre = open( "textes/missy/cendre.txt", encoding='utf-8' ).read()

collation.add_plain_witness( "MS4A3a", temoin_MS4A3a )
collation.add_plain_witness( "MS4A1a", temoin_MS4A1a )
collation.add_plain_witness( "Cendre", temoin_cendre )

table = collate(collation, layout='vertical', output='html2')
print(table)
```

```python
from collatex import *

collation = Collation()

temoin_MS4A3a = open( "textes/missy/MS4A3a.txt", encoding='utf-8' ).read()
temoin_MS4A1a = open( "textes/missy/MS4A1a.txt", encoding='utf-8' ).read()
temoin_cendre = open( "textes/missy/cendre.txt", encoding='utf-8' ).read()


collation.add_plain_witness( "MS4A3a", temoin_MS4A3a )
collation.add_plain_witness( "MS4A1a", temoin_MS4A1a )
collation.add_plain_witness( "Cendre", temoin_cendre )

table = collate(collation, layout='vertical', near_match=True, output='html2', segmentation=False)
print(table)
```
